```julia
using Revise
using DTWA
using DiffEqBase
using Plots

plotly()
```

```julia
J = 1 * 2π * [0.0 1.0
     1.0 0.0];
h = 2π * [0.0, 0.0, 0.0];
anisotropy = [0.0, 1.0, 0.0]
system = XYZSystem(J, h, anisotropy)
Ψ = [1., 0., 0.]
Ψ = ProductState(2, Ψ)
times = LinRange(0, 2, 100)
problem = DTWAProblem(system, Ψ, times)
```

```julia
sol = solve_dtwa(problem, 1000);
4
```

```julia
plot(times[2:end], 0.5 * sol[2:end, 1], label="dtwa")
xaxis!("time [μs]")
yaxis!("magnetization")
```
