```julia
J = 0 * 2π * [0.0 1.0
     1.0 0.0];
N = size(J)[1]
h = 2π * 1 * [0, 1.0, 0.0];
anisotropy = [0.0, 0.0, 1.0]
system = XYZSystem(J, h, anisotropy)
Ψ = [1., 0., 0.]
Ψ = ProductState(N, Ψ)
times = LinRange(0, 4, 100)
problem = DTWAProblem(system, Ψ, times)
```

```julia
sol = solve_dtwa(problem, 1000);
4
```
